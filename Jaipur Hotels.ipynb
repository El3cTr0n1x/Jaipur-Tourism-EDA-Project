```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
Hotels_data = pd.read_csv('/Users/abhaykatti/Desktop/Data Analytics/Projects/Jaipur Tourism/Hotels.csv')
Places_data = pd.read_csv('/Users/abhaykatti/Desktop/Data Analytics/Projects/Jaipur Tourism/Places.csv')
Types_data = pd.read_csv('/Users/abhaykatti/Desktop/Data Analytics/Projects/Jaipur Tourism/Types.csv')
```

```python
Hotels_df = hotels_data.drop(['Lat','Lng','HID'], axis = 1)
```

```python
Hotels_df.info()
```

```python
Hotels_df.describe()
```

```python
cut_labels = ['Low','Moderate','High']
Hotels_df['Budget'] = pd.qcut(Hotels_df['PRICE_RUPEES'][Hotels_df['PRICE_RUPEES']>0],q=[0,.25, .75, 1], labels=cut_labels).astype("object")
Hotels_df['Popularity'] = pd.qcut(Hotels_df['NUMBER_OF_REVIEWS'][Hotels_df['NUMBER_OF_REVIEWS']>0],q=[0,.25, .75, 1], labels=cut_labels).astype("object")
```

```python
Hotels_df
```

```python
sbn.displot(x= Hotels_df['Budget'], hue= Hotels_df['Popularity'] , palette = 'terrain_r')
plt.show()
```

```python
Hotels_df['Budget'].value_counts()
```

```python
Hotels_df.groupby(['Budget', 'Popularity'], as_index=False)[['PRICE_RUPEES', 'NUMBER_OF_REVIEWS']].agg(['mean', 'sum', 'count', 'median', 'min', 'max'])
```

```python
sbn.catplot(x='PRICE_RUPEES', y= "NUMBER_OF_REVIEWS", data = Hotels_df.head(7), kind = 'bar')
plt.show()
```

```python
sbn.violinplot(x= 'Budget', y = 'NUMBER_OF_REVIEWS', data =Hotels_df,hue= "Popularity",palette = 'plasma')
plt.show()
```
